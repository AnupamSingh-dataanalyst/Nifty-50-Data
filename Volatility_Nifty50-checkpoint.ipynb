```python
# For this we have to import below libraries in Python Jupyter Notebook
# Below i added print command which notify me once library get imported 
import pandas as pd
import wget
import numpy as np
print(" setup complete")
```

```python
#download raw data set 
#url of the raw csv dataset from NSE website 

urls = ['https://www1.nseindia.com/archives/nsccl/volt/CMVOLT_08022021.CSV']

[wget.download(url) for url in urls]
```

```python
#As we can see above our csv file is downloaded as output.Now it is time to convert this in dataframe to read the file

DV_df = pd.read_csv('CMVOLT_08022021.CSV')
```

```python
#Just to check  the columns 
DV_df.columns
```

```python
# To set symbols as index column. This helps to do furthur operations on symbol.
DV_df.set_index('Symbol', inplace=True )
```

```python
# Now time for  some data wrangling. Convert naming  Daily Volatility & Annual Volatility Columns to A & B for easyness.
DV_df = DV_df.rename({"Current Day Underlying Daily Volatility (E) = Sqrt(0.995*D*D + 0.005*C*C)":"A","Underlying Annualised Volatility (F) = E*Sqrt(365)":"B",}, axis='columns')
DV_df.head()
```

```python
# One more step before our final Destination . Just Add Symbols name(of our Interest ) in row and Volatility in Columns 
rows=['ADANIPORTS','EICHERMOT','BANKBARODA','CIPLA','BAJAJ-AUTO','ASIANPAINT','COALINDIA','BHARTIARTL','BRITANNIA','GAIL','KOTAKBANK','ONGC','TATAMOTORS','UPL','IOC','HINDALCO','GRASIM','COALINDIA','RELIANCE','SBIN','AXISBANK','BPCL']
columns=['A','B']
NF=DV_df.loc[rows,columns]
```

```python
# Hurray! Now every time in a blink of an eye we can check EOD Nifty 50 Volatility just by running
#this Jupyter Notebook File. Happy Stcok Trading.

NF.sort_values(by="A",ascending=False)
```
